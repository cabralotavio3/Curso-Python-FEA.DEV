**Exercício 1**

```python
# list initialization
teams = ["Palmeiras", "Vasco", "Cruzeiro", "Flamengo"]

# a) adding Santos
teams.append("Santos")

# b) removing the second team (index 1)
disqualified = teams.pop(1)

# c) checking membership
has_corinthians = "Corinthians" in teams

# d) printing results
print(f"Time desclassificado: {disqualified}")
print(f"Lista atualizada: {teams}")
print(f"Tamanho da lista: {len(teams)}")

if has_corinthians:
    print("O Corinthians está no torneio.")
else:
    print("O Corinthians não está no torneio.")
```

**Exercício 2**

```python
# variables setup
stock_quantity = 50
raining = False
cold = True

# a) checking if stock is not empty 
if stock_quantity:
    # b) checking weather conditions using OR
    if raining or cold:
        print("Restaurante aberto e focado no delivery!")
    # c) fallback if not raining/cold
    else:
        print("Restaurante aberto para o público!")
else:
    # empty stock (0)
    print("Restaurante fechado!")
```

**Exercício 3**

```python
# variables setup
phrase = "Palmeiras nao tem mundial"
count_a = 0

# a) and b) loop through each character
for char in phrase:
    # c) checking if char is 'a'
    if char == "a":
        count_a += 1

# d) printing results
print(f"Total de letras 'a' encontradas: {count_a}")
```

**Exercício 4**

```python
num = 10
original_num = num
factorial = 1

# a) while loop for factorial
while num > 0:
    # b) safety break for numbers larger than 15
    if original_num > 15:
        print("Aviso: Número muito grande, cálculo interrompido.")
        break
    
    factorial *= num
    num -= 1

# c) printing if finished normally
if original_num <= 15:
    print(f"{original_num}! = {factorial}")
```
