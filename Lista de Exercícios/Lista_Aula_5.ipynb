### Gabarito — Exercício 4

````python
v = np.array([1, 2, 3])

norma = np.linalg.norm(v)
identidade = np.eye(3)
resultado  = np.matmul(identidade, v)

print("norma:", norma)
print("I @ v:", resultado)
print("igual a v?", np.allclose(resultado, v))
````

### Gabarito — Exercício 9

````python
print("Antes:")
print(df[["duracao_min", "ano"]].dtypes)

df["duracao_min"] = df["duracao_min"].astype(float)
df["ano"]         = df["ano"].astype(str)

print("\nDepois:")
print(df[["duracao_min", "ano"]].dtypes)

# Revertendo para o tipo original para não quebrar exercícios seguintes
df["ano"] = df["ano"].astype(int)
````

### Gabarito — Exercício 10

````python
import numpy as np

print("=== nota_imdb > 8.0 ===")
print(df[df["nota_imdb"] > 8.0][["titulo", "nota_imdb"]])

print("\n=== Terror ===")
print(df[df["genero"] == "Terror"][["titulo", "genero"]])

mask = np.bitwise_and(df["nota_imdb"] > 7.5, df["bilheteria_mi"] > 100)
print("\n=== nota > 7.5 E bilheteria > 100 ===")
print(df[mask][["titulo", "nota_imdb", "bilheteria_mi"]])
````
