## **Gabarito**

#### Exercício 1

```python
usd_texto = input("Digite a quantidade de Dólares (USD) que você possui: ")
cotacao_texto = input("Digite a cotação atual do Dólar em Reais (BRL): ")

# convertendo as strings do input para float
usd = float(usd_texto)
cotacao = float(cotacao_texto)

# calculando em reais
brl = usd * cotacao

# exibindo o resultado formatado com 2 casas decimais
print(f"Seus USD {usd:.2f} equivalem a R$ {brl:.2f} na cotação atual.")
```

#### Exercício 2

```python
valor_inicial = float(input("Digite o Valor Inicial do Investimento (R$): "))
valor_final = float(input("Digite o Valor Final do Investimento (R$): "))

# calculando o ROI
roi = ((valor_final - valor_inicial) / valor_inicial) * 100

print(f"O Retorno Sobre Investimento (ROI) foi de {roi:.2f}%")
```
