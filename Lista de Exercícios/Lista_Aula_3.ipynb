**`append(item)`** adiciona **um único item** ao final da lista — se o item for uma lista `[4, 5]`, ela é inserida como um elemento só (a lista fica com um elemento a mais que é uma sub-lista).

## Tópico 2 - Tuplas


### Gabarito 2.1 - Teórica

Tuplas podem ser chaves de dicionários porque são **imutáveis**: após a criação, seus elementos não podem ser alterados, o que garante que o valor do hash permaneça constante durante toda a vida do objeto. Dicionários e sets exigem que suas chaves sejam *hashable*, e objetos mutáveis (como listas) não podem ser hashados porque mudar o conteúdo mudaria o hash, corrompendo a tabela interna.

Exemplo prático: um **horário de voo** `("GRU", "GIG", "06:30")` — origem, destino e horário são dados fixos que identificam um voo, nunca devem mudar, e faz sentido usá-los como chave em um dicionário de preços.


### Gabarito 2.2 - Prática


```python
def estatisticas(numeros):
    maior = max(numeros)
    menor = min(numeros)
    media = sum(numeros) / len(numeros)
    return maior, menor, media  # retorna tupla implicitamente

notas = [6.5, 9.0, 4.5, 8.0, 7.5, 10.0, 3.0]

# Desempacotamento de tupla
maior, menor, media = estatisticas(notas)
print(f"Maior nota:  {maior}")
print(f"Menor nota:  {menor}")
print(f"Média:       {media:.2f}")
```

### Gabarito 2.3 - Prática


```python
# Parte 1: tuplas como chaves de dicionário
distancias = {
    ("São Paulo", "Rio de Janeiro"): 430,
    ("São Paulo", "Belo Horizonte"): 586,
    ("Rio de Janeiro", "Salvador"): 1649,
}

print("Dicionário de distâncias:", distancias)
print("SP → RJ:", distancias[("São Paulo", "Rio de Janeiro")], "km")

# Parte 2: desempacotamento com *
sequencia = (1, 2, 3, 4, 5, 6, 7)

primeiro, *meio, ultimo = sequencia
print("Primeiro:", primeiro)   # 1
print("Meio:", meio)           # [2, 3, 4, 5, 6]
print("Último:", ultimo)      # 7
```

## Tópico 3 - Conjuntos (Sets)


### Gabarito 3.1 - Teórica

Quando um elemento é adicionado a um set, Python aplica a função `hash()` sobre ele, obtendo um número inteiro. Esse número determina em qual "balde" (bucket) da memória o elemento será armazenado. Na verificação `x in conjunto`, Python calcula `hash(x)` e vai **diretamente** ao balde correspondente — sem percorrer toda a coleção. O custo é aproximadamente constante, independentemente do tamanho do conjunto (**O(1)**).

Sets não aceitam duplicatas porque, antes de inserir, verificam se já existe um elemento com o mesmo hash **e** igual ao novo; se sim, ignoram a inserção.

A ausência de ordem garantida ocorre porque a posição de cada elemento é determinada pelo seu hash, não pela ordem de inserção — elementos diferentes podem cair em buckets em qualquer sequência.


### Gabarito 3.2 - Prática


```python
janeiro   = [101, 102, 103, 104, 105, 106]
fevereiro = [103, 105, 107, 108, 101, 109]

jan = set(janeiro)
fev = set(fevereiro)

# 1. Todos os clientes (união)
todos = jan | fev
print("Todos os clientes:", todos)

# 2. Clientes fidelizados (interseção)
fidelizados = jan & fev
print("Compraram nos dois meses:", fidelizados)

# 3. Apenas em janeiro (diferença)
so_janeiro = jan - fev
print("Apenas em janeiro:", so_janeiro)

# 4. Em apenas um dos meses (diferença simétrica)
so_um_mes = jan ^ fev
print("Em apenas um mês:", so_um_mes)
```

### Gabarito 3.3 - Prática


```python
texto = """
Python e uma linguagem de programacao muito usada em ciencia de dados.
Com Python voce pode analisar dados, criar modelos e automatizar tarefas.
Python e simples e poderosa ao mesmo tempo.
"""

# 1. Criar o set de palavras únicas (normalizado)
palavras = texto.lower().replace('.', '').replace(',', '').split()
unicas = set(palavras)

# 2. Quantidade de palavras únicas
print(f"Palavras únicas: {len(unicas)}")
print(unicas)

# 3. Verificação de pertencimento
for palavra in ['python', 'java', 'dados']:
    print(f"'{palavra}' está no texto? {palavra in unicas}")
```

### Gabarito 4.1 - Teórica

`dicionario["chave"]` lança um **`KeyError`** se a chave não existir. `.get("chave", valor_padrão)` retorna o valor padrão informado (ou `None` se omitido) sem lançar exceção — é preferível quando se trabalha com dados externos, APIs ou qualquer situação em que a ausência de uma chave é esperada e não deve interromper o programa.

Internamente, ao buscar `dicionario[chave]`, Python: (1) calcula `hash(chave)`; (2) usa esse hash para localizar o bucket na tabela hash; (3) compara a chave encontrada com a chave buscada (via `==`). Se coincidirem, retorna o valor. O custo médio é **O(1)**.


### Gabarito 4.2 - Prática


```python
estoque = {
    "maca": 80,
    "banana": 45,
    "uva": 20,
    "laranja": 60,
    "manga": 35
}

# 1. Adicionar abacate
estoque["abacate"] = 40
print("Após adicionar abacate:", estoque)

# 2. Atualizar banana
estoque["banana"] = 90
print("Após atualizar banana:", estoque)

# 3. Remover uva
qtd_uva = estoque.pop("uva")
print(f"Uva removida (quantidade era {qtd_uva}):", estoque)

# 4. Produtos com quantidade abaixo de 50
print("\nProdutos com menos de 50 unidades:")
for produto, qtd in estoque.items():
    if qtd < 50:
        print(f"  {produto}: {qtd}")
```

### Gabarito 4.3 - Prática


```python
catalogo = {
    "eletronicos": {
        "Notebook": 3500.0,
        "Smartphone": 2200.0,
        "Tablet": 1800.0
    },
    "livros": {
        "Python para Todos": 89.90,
        "Ciencia de Dados do Zero": 120.00,
        "Algoritmos": 150.00
    },
    "vestuario": {
        "Camiseta": 59.90,
        "Calca Jeans": 189.90,
        "Tenis": 349.90
    }
}

# 1. Nome e preço de todos os produtos
print("=== Catálogo Completo ===")
for categoria, produtos in catalogo.items():
    print(f"\n{categoria.upper()}:")
    for nome, preco in produtos.items():
        print(f"  {nome}: R${preco:.2f}")

# 2. Preço médio por categoria
print("\n=== Preço Médio por Categoria ===")
for categoria, produtos in catalogo.items():
    precos = list(produtos.values())
    media = sum(precos) / len(precos)
    print(f"  {categoria}: R${media:.2f}")

# 3. Produto mais caro do catálogo
mais_caro_nome = None
mais_caro_cat  = None
mais_caro_preco = 0

for categoria, produtos in catalogo.items():
    for nome, preco in produtos.items():
        if preco > mais_caro_preco:
            mais_caro_preco = preco
            mais_caro_nome  = nome
            mais_caro_cat   = categoria

print(f"\nProduto mais caro: {mais_caro_nome} "
      f"({mais_caro_cat}) — R${mais_caro_preco:.2f}")
```

## Tópico 5 - Arrays


### Gabarito 5.1 - Teórica

Uma lista Python guarda **ponteiros** para objetos Python independentes na memória. Cada objeto carrega metadados (tipo, contagem de referências, etc.), o que consome muito mais espaço do que o valor bruto.

Um `array` (ou array NumPy) armazena os valores **diretamente e contiguamente** na memória, sem envoltório de objeto Python para cada número. A restrição que viabiliza isso é que **todos os elementos devem ser do mesmo tipo** (definido pelo typecode). Isso elimina a necessidade de ponteiros e metadados individuais.

Situação prática: processamento de grandes séries temporais de preços (milhões de floats), leitura de dados binários de sensores, ou qualquer tarefa que precise de baixo consumo de memória e operações numéricas rápidas.


### Gabarito 5.2 - Prática


```python
import array
import sys

# 1. Criar lista e array com 500 inteiros
lista_nums = list(range(500))
array_nums = array.array('i', range(500))

# 2. Tamanho em bytes
mem_lista = sys.getsizeof(lista_nums)
mem_array = sys.getsizeof(array_nums)
print(f"Lista: {mem_lista} bytes")
print(f"Array: {mem_array} bytes")

# 3. Diferença percentual
excesso = (mem_lista - mem_array) / mem_array * 100
print(f"A lista ocupa {excesso:.1f}% a mais de memória que o array.")

# 4. Tentativa de adicionar float num array de inteiros
try:
    array_nums.append(3.7)
except TypeError as e:
    print(f"\nErro ao inserir 3.7: {e}")
    # O array rejeita o float porque foi criado com typecode 'i' (inteiro 32-bit).
    # Ele só aceita valores do tipo declarado na criação.
```

### Gabarito 5.3 - Prática


```python
import numpy as np

celsius = [0, 20, 37, 100, -10, 25]

# 1. Conversão vetorizada para Fahrenheit
arr_c = np.array(celsius)
arr_f = arr_c * 9/5 + 32
print("Celsius:    ", arr_c)
print("Fahrenheit: ", arr_f)

# 2. Matriz 3x3 de 1 a 9
matriz = np.arange(1, 10).reshape(3, 3)
print("\nMatriz 3x3:")
print(matriz)
print("Elemento [1][2] (linha 2, coluna 3):", matriz[1, 2])  # 6

# 3. Estatísticas das temperaturas em Celsius
print(f"\nMédia:  {arr_c.mean():.2f} °C")
print(f"Máximo: {arr_c.max()} °C")
print(f"Mínimo: {arr_c.min()} °C")
```

| **Módulo `csv`** | Leitura/escrita de arquivos CSV sem dependências externas, scripts simples de ETL |
| **Pandas DataFrame** | Análise exploratória, grandes volumes, operações de filtro/agrupamento/join, integração com bibliotecas de ML |

**Limitações da lista de dicionários em grandes volumes:** cada operação (filtro, ordenação, cálculo de média) exige um loop Python explícito, que é lento. Não há vetorização, tipagem eficiente ou operações de agrupamento nativas. Com milhões de registros, o consumo de memória também se torna muito maior do que um DataFrame ou array NumPy.


### Gabarito 6.2 - Prática


```python
vendedores = [
    {"nome": "Alice",   "regiao": "Sul",     "vendas": 45000, "meta": 40000},
    {"nome": "Bob",     "regiao": "Norte",   "vendas": 32000, "meta": 38000},
    {"nome": "Carla",   "regiao": "Sudeste", "vendas": 61000, "meta": 55000},
    {"nome": "Diego",   "regiao": "Nordeste","vendas": 28000, "meta": 30000},
    {"nome": "Elisa",   "regiao": "Centro",  "vendas": 53000, "meta": 50000},
]

# 1. Vendedores com meta batida
print("Meta batida:")
for v in vendedores:
    if v["vendas"] >= v["meta"]:
        print(f"  {v['nome']} — R${v['vendas']:,} (meta: R${v['meta']:,})")

# 2. Total de vendas
total = sum(v["vendas"] for v in vendedores)
print(f"\nTotal de vendas: R${total:,}")

# 3. Ranking por vendas (maior → menor)
ranking = sorted(vendedores, key=lambda v: v["vendas"], reverse=True)
print("\nRanking:")
for pos, v in enumerate(ranking, start=1):
    print(f"  {pos}. {v['nome']}: R${v['vendas']:,}")
```

### Gabarito 6.3 - Prática


```python
import csv
import os

vendedores = [
    {"nome": "Alice",   "regiao": "Sul",     "vendas": 45000, "meta": 40000},
    {"nome": "Bob",     "regiao": "Norte",   "vendas": 32000, "meta": 38000},
    {"nome": "Carla",   "regiao": "Sudeste", "vendas": 61000, "meta": 55000},
    {"nome": "Diego",   "regiao": "Nordeste","vendas": 28000, "meta": 30000},
    {"nome": "Elisa",   "regiao": "Centro",  "vendas": 53000, "meta": 50000},
]

arquivo = "vendedores.csv"

# 1. Salvar como CSV
with open(arquivo, "w", newline="", encoding="utf-8") as f:
    campos = ["nome", "regiao", "vendas", "meta"]
    writer = csv.DictWriter(f, fieldnames=campos)
    writer.writeheader()
    writer.writerows(vendedores)
print(f"Arquivo '{arquivo}' criado.")

# 2. Ler o CSV e reconstruir com tipos corretos
reconstruido = []
with open(arquivo, "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for linha in reader:
        reconstruido.append({
            "nome":   linha["nome"],
            "regiao": linha["regiao"],
            "vendas": int(linha["vendas"]),
            "meta":   int(linha["meta"]),
        })

# 3. Imprimir e confirmar tipos
for v in reconstruido:
    print(v)
print(f"\nTipo de 'vendas': {type(reconstruido[0]['vendas'])}")
print(f"Tipo de 'meta':   {type(reconstruido[0]['meta'])}")

# 4. Deletar o arquivo
os.remove(arquivo)
print(f"\nArquivo '{arquivo}' removido.")
```
