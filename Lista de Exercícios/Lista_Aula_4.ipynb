```python
def calculate_power(*stats: int, level:int = 1) -> float:
    if not stats:
        return 0
    return (sum(stats) / len(stats)) * level

def create_character(name, **attributes):
    print(f"\n>>> Personagem: {name} <<<")
    for attr, value in attributes.items():
        print(f"  {attr}: {value}")

create_character("Gandalf", strength = 5, agility = 3, intelligence = 10, magic = 9)
gandalf_power = calculate_power(5, 3, 10, 9, level = 5)
print(f"[Poder Total] -> de Gandalf: {gandalf_power:.1f}")

create_character("Aragorn", strength = 9, agility = 8, intelligence = 6, magic=2)
aragorn_power = calculate_power(9, 8, 6, 2, level = 3)
print(f"[Poder Total] -> de Aragorn: {aragorn_power:.1f}")

if gandalf_power > aragorn_power:
    print("\nResultado do Combate: Gandalf é mais poderoso!")
else:
    print("\nResultado do Combate: Aragorn é mais poderoso!")
```

```python
# ord('a') = 97 

def encrypt(text: str, shift: int) -> str:
    result = ""
    for char in text:
        if char.islower():
            new_pos = (ord(char) - ord("a") + shift) % 26
            result += chr(new_pos + ord("a"))
        else:
            result += char
    return result

def decrypt(text: str, shift: int) -> str:
    return encrypt(text, -shift)

def process_message(text: str, shift: int, operation) -> str:
    return operation(text, shift)

original = "python incrivel"
encrypted = process_message(original, 3, encrypt)
decrypted = process_message(encrypted, 3, decrypt)

print(f"Texto Original:     {original}")
print(f"   Texto Cifrado:       {encrypted}")
print(f"   Texto Decifrado:   {decrypted}")
```

**Exercício 14**

```python
class Student:
    def __init__(self, name: str, grades: list[float]) -> None:
        self.name = name
        self.grades = grades

    def average(self) -> float:
        total = 0
        for grade in self.grades:
            total += grade
        return total / len(self.grades)

    def status(self) -> str:
        if self.average() >= 7.0:
            return "Aprovado"
        return "Reprovado"

s1 = Student("Carlos", [8.0, 6.5, 9.0])
s2 = Student("Maria", [5.0, 4.5, 6.0])

for student in [s1, s2]:
    print(f"Aluno: {student.name} — Média: {student.average():.1f} — {student.status()}")
```

**Exercício 15**

```python
class BankAccount:
    def __init__(self, holder: str, balance: float = 0.0):
        self.holder = holder
        self.balance = balance

    def deposit(self, amount: float) -> None:
        if amount < 0:
            print("[Erro] Valor inválido para depósito")
            return
        self.balance += amount

    def withdraw(self, amount: float) -> None:
        if amount > self.balance:
            print("[Erro] Saldo insuficiente para saque")
            return
        self.balance -= amount

    def statement(self) -> None:
        print(f"Extrato -> Titular: {self.holder} | Saldo: R$ {self.balance:.2f}")

account = BankAccount("João")
account.deposit(500)
account.withdraw(200)
account.withdraw(400)
account.statement()
```

**Exercício 16**

```python
class Track:
    def __init__(self, title: str, artist: str, duration: float) -> None:
        self.title = title
        self.artist = artist
        self.duration = duration


class Playlist:
    def __init__(self, name: str) -> None:
        self.name = name
        self.tracks = []

    def add_track(self, track: Track) -> None:
        self.tracks.append(track)

    def total_duration(self) -> float:
        total = 0
        for track in self.tracks:
            total += track.duration
        return total

    def average_duration(self) -> float:
        if not self.tracks:
            return 0.0
        return self.total_duration() / len(self.tracks)

    def longest_track(self) -> Track:
        longest = self.tracks[0]
        for track in self.tracks:
            if track.duration > longest.duration:
                longest = track
        return longest


road_trip = Playlist("Road Trip")
road_trip.add_track(Track("Bohemian Rhapsody", "Queen", 5.92))
road_trip.add_track(Track("Hotel California", "Eagles", 6.51))
road_trip.add_track(Track("Stairway to Heaven", "Led Zeppelin", 8.03))
road_trip.add_track(Track("Sweet Child O' Mine", "Guns N' Roses", 5.56))

print(f"Duração total da Playlist: {road_trip.total_duration():.2f} min")
print(f"   Duração média por faixa: {road_trip.average_duration():.2f} min")
print(f"   Faixa mais longa: {road_trip.longest_track().title}")
```

```python
import time

def measure_time(func, *args):
    start = time.perf_counter()
    result = func(*args)
    end = time.perf_counter()
    return (result, end - start)

def sum_range(n: int) -> int:
    total = 0
    for i in range(n):
        total += i
    return total

result, elapsed = measure_time(sum_range, 1_000_000)
print(f"Resultado da soma: {result}")
print(f"   Tempo de execução: {elapsed:.6f}s")
```

```python
def analyze_password(password: str) -> tuple[int, int, int, int]:
    length = len(password)
    uppercase = 0
    digits = 0
    specials = 0

    for char in password:
        if char.isupper():
            uppercase += 1
        elif char.isdigit():
            digits += 1
        elif not char.isalpha():
            specials += 1

    return (length, uppercase, digits, specials)


def calculate_score(*counts: int) -> int:
    length, uppercase, digits, specials = counts
    return length * 2 + uppercase * 3 + digits * 4 + specials * 5


def classify_strength(score: int, **thresholds: int) -> str:
    weak = thresholds.get("weak", 20)
    medium = thresholds.get("medium", 40)
    strong = thresholds.get("strong", 60)

    if score < weak:
        return "Fraca"
    elif score < medium:
        return "Média"
    elif score < strong:
        return "Forte"
    else:
        return "Muito forte"


passwords = ["abc123", "Python2026!", "S3nh@_F0rt3#2026"]

for pwd in passwords:
    length, upper, digs, specs = analyze_password(pwd)
    score = calculate_score(length, upper, digs, specs)
    strength = classify_strength(score)
    print(f"Senha: '{pwd}' -> -> Score: {score} | Classificação: {strength}")
```

```python
import math


class City:
    def __init__(self, name: str, latitude: float, longitude: float) -> None:
        self.name = name
        self.latitude = latitude
        self.longitude = longitude


def haversine_distance(city_a: City, city_b: City) -> float:
    r = 6371  # Earth radius in km

    lat1 = math.radians(city_a.latitude)
    lat2 = math.radians(city_b.latitude)
    
    dlat = math.radians(city_b.latitude - city_a.latitude)
    dlon = math.radians(city_b.longitude - city_a.longitude)

    a = math.sin(dlat / 2) ** 2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon / 2) ** 2
    d = 2 * r * math.atan2(math.sqrt(a), math.sqrt(1 - a))

    return d


def flight_cost(distance_km: float, **kwargs: float) -> float:
    base_fare = kwargs.get("base_fare", 150.0)
    price_per_km = kwargs.get("price_per_km", 0.45)
    return base_fare + price_per_km * distance_km


def plan_trip(*cities: City) -> None:
    total_distance = 0
    total_cost = 0

    for i in range(len(cities) - 1):
        origin = cities[i]
        destination = cities[i + 1]

        dist = haversine_distance(origin, destination)
        cost = flight_cost(dist)

        total_distance += dist
        total_cost += cost

        print(f"Trecho: {origin.name} -> {destination.name}: {dist:.1f} km | R$ {cost:.2f}")

    print(f"\nDistância total: {total_distance:.1f} km")
    print(f"Custo total: R$ {total_cost:.2f}")


# Test
sp = City("São Paulo", -23.55, -46.63)
bsb = City("Brasília", -15.79, -47.88)
ssa = City("Salvador", -12.97, -38.51)
rec = City("Recife", -8.05, -34.87)

plan_trip(sp, bsb, ssa, rec)
```
