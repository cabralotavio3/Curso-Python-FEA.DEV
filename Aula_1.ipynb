## 1. Variáveis, Tipos de Dados Fundamentais e Palavras Reservadas

### O que são Variáveis?
Pense em uma **variável** como uma caixa onde você guarda uma informação. Você dá um nome a essa caixa para poder acessar o dado mais tarde.

```python
# Criando variável | nome_da_variavel = valor_recebido
nome = "Gustavo"
idade = 18
```

### Tipos de Dados Fundamentais
Em Python, cada valor tem um "tipo". Os quatro tipos principais são:

* **int:** Números inteiros (ex: 10, -5)
* **float:** Números com ponto flutuante/decimais (ex: 10.5, 3.14)
* **str:** Strings, ou cadeias de caracteres (texto) (ex: "Olá", 'Python')
* **bool:** Booleanos, valores lógicos (True ou False)

### Casting (Conversão de Tipo)
Muitas vezes, recebemos um dado em um formato, mas precisamos dele em outro. "Casting" é o ato de forçar um dado a mudar de tipo. Por exemplo, transformar o texto `"10"` no número inteiro `10` usando a função `int()`.

```python
# Verificando tipos de dados com a função type()
print(type(10))         # int
print(type(3.14))       # float
print(type("Python"))   # str
print(type(True))       # bool

# Exemplo de tipagem dinâmica (o tipo da variável pode mudar)
variavel = 50
print(f"O valor é {variavel} e o tipo é {type(variavel)}")

variavel = "Agora sou um texto"
print(f"O valor é {variavel} e o tipo é {type(variavel)}")

# Observação: a tipagem dinâmica não ocorre em todas as linguagens de programação

# Exemplo de Casting (Conversão)
numero_em_texto = "50"
print(f"Tipo antes da conversão: {type(numero_em_texto)}")

# Convertendo string para inteiro
numero_inteiro = int(numero_em_texto)
print(f"Tipo depois da conversão: {type(numero_inteiro)}")
```

## 2. Operadores Aritméticos

### Operações Básicas
Python funciona como uma calculadora poderosa. Aqui estão os operadores essenciais:

* `+  : Adição`
* `-  : Subtração`
* `*  : Multiplicação`
* `/  : Divisão (retorna sempre float)`
* `// : Divisão Inteira (descarta a parte decimal)`
* `%  : Módulo (resto da divisão)`
* `** : Exponenciação (potência)`

```python
a = 10
b = 3

# Operações
print(f"Adição: {a} + {b} = {a + b}")
print(f"Subtração: {a} - {b} = {a - b}")
print(f"Multiplicação: {a} * {b} = {a * b}")

# Note a diferença entre as divisões
print(f"Divisão Real: {a} / {b} = {a / b}")
print(f"Divisão Inteira: {a} // {b} = {a // b}")

# Módulo e Potência
print(f"Resto da divisão (Módulo): {a} % {b} = {a % b}")
print(f"Potência (10 elevado a 3): {a} ** {b} = {a ** b}")
```

### Ordem de Precedência (PEMDAS)
Assim como na matemática, Python resolve expressões na seguinte ordem: Parênteses, Expoentes, Multiplicação/Divisão, Adição/Subtração.

```python
# Ordem de precedência (PEMDAS)
# 1. Parênteses
# 2. Expoentes
# 3. Multiplicação e Divisão
# 4. Adição e Subtração

resultado = 10 + 5 * 2  # Multiplicação acontece primeiro
print(f"10 + 5 * 2 = {resultado}")

resultado_com_parenteses = (10 + 5) * 2 # Parênteses alteram a ordem
print(f"(10 + 5) * 2 = {resultado_com_parenteses}")
```

## 3. Operações com Strings

### Concatenação e Repetição
Strings são sequências de caracteres. Podemos manipulá-las usando operadores matemáticos para concatenar (juntar com `+`) ou repetir (com `*`).

```python
texto1 = "Data"
texto2 = "Science"

# Concatenação (+)
curso = texto1 + " " + texto2
print(curso)

# Repetição (*)
divisoria = "-" * 30
print(divisoria)
print("Atenção!")
print(divisoria)
```

### Por que usamos `.algo()`? (Métodos)
Em Python, quase tudo é um **Objeto**. Quando colocamos um ponto depois de uma variável (ex: `texto.upper()`), estamos chamando um **método**. Um método é uma ação específica que aquele tipo de dado sabe fazer. Como, por exemplo: strings têm métodos para ficar em maiúsculo, já inteiros não!

```python
# Manipulação básica de Strings (Métodos)
frase = "  Python é Incrível  "

print(f"Original: '{frase}'")
print(f"Maiúsculas: {frase.upper()}")
print(f"Minúsculas: {frase.lower()}")
print(f"Sem espaços nas pontas: '{frase.strip()}'")
print(f"Contagem de letras 'i': {frase.lower().count('i')}")
```

### Formatação de Strings (f-strings)
Uma das ferramentas mais úteis do Python é a **f-string**. Ela permite misturar texto e variáveis de forma limpa, sem precisar usar o sinal de `+` várias vezes. Basta colocar um `f` antes das aspas e usar chaves `{}` para inserir a variável.

```python
# Formatação de valores em f-strings
lucro = 1450.876543
print(f"O lucro foi de R$ {lucro:.2f}") # .2f formata para 2 casas decimais

taxa = 0.154
print(f"A taxa foi de {taxa:.1%}") # .1% formata como porcentagem

nome = "Maria"
idade = 25

# Sem f-string (trabalhoso e propenso a erros de espaço)
print("A aluna " + nome + " tem " + str(idade) + " anos.")

# Com f-string (mais limpo e converte tipos automaticamente)
print(f"A aluna {nome} tem {idade} anos.")
```

## 4. Inputs e Outputs

Para tornar nossos programas interativos, precisamos receber dados do usuário (`input`) e mostrar resultados na tela (`print`).

**Importante:** A função `input()` sempre retorna o dado como uma **string**. Se precisarmos fazer contas, devemos converter para `int` ou `float`.

```python
# Output simples
print("Este é um output simples na tela.")

# Input simples
nome = input("Digite seu nome: ")
print(f"Olá, {nome}! Bem-vindo ao curso.")
```

```python
# Input com conversão de tipos
numero1 = input("Digite o primeiro número: ")
numero2 = input("Digite o segundo número: ")

# Convertendo as strings recebidas para inteiros
soma = int(numero1) + int(numero2)

print(f"A soma de {numero1} com {numero2} é igual a {soma}")

# Outra forma para converter um input como inteiro
x = int(input("Digite o valor de x:"))

print(type(x))
print(x)
```

### O Erro Mais Comum: TypeError

Lembre-se: O `input` sempre lê texto (string). Se tentarmos somar uma string com um número sem converter antes, o Python gerará um erro.

```python
# Exemplo de ERRO
# O usuário digita 5, mas o Python lê "5" (texto)
numero = input("Digite um número para o erro: ") 

# A linha abaixo vai falhar porque não podemos somar Texto + Número
# O Python não sabe se você quer somar matemática ou colar texto.

# A forma incorreta:

# print(numero + 10) 

# A forma correta seria:
print(int(numero) + 10)
```

## 5. Capturando Exceções (Erros) com Try e Except

Nem sempre as coisas saem como planejado. O que acontece se tentarmos dividir um número por zero? Ou se tentarmos dividir por um texto? O programa "quebra" gerando uma **Exceção**.

```python
# Se executarmos a linha abaixo, o Python dispara um erro (ZeroDivisionError)
# print(10 / 0)
```

### Evitando que o código quebre
Usamos o bloco `try` e `except`. Ele significa: *"Tente executar isso. Se der um erro específico, faça esta outra coisa em vez de parar o programa"*.

```python
try:
    numero = int(input("Digite um número: "))
    print(10 / numero)
except ValueError:
    print("Erro: Digite um valor numérico válido!")
except ZeroDivisionError:
    print("Erro: Não se pode dividir por zero!")
```

### Descobrindo o tipo do Input
Podemos usar o `try/except` para tentar converter o que o usuário digitou (casting) repetidamente até descobrir qual era o tipo real daquele dado.

```python
entrada = input("Digite qualquer coisa: ")

try:
    valor = int(entrada)
    print(f"Você digitou um Inteiro (int): {valor}")
except ValueError:
    try:
        valor = float(entrada)
        print(f"Você digitou um Decimal (float): {valor}")
    except ValueError:
        if entrada.lower() == 'true' or entrada.lower() == 'false':
            print(f"Você digitou um Booleano (bool): {entrada}")
        else:
            print(f"Você digitou uma String (str): '{entrada}'")
```

## 6. Exemplos Práticos de Aplicação

Vamos juntar tudo o que aprendemos (inputs, conversão de tipos, operações matemáticas, métodos e formatação de strings) para resolver problemas reais.

### Exemplo 1: Calculadora de Juros Compostos
Calculando o retorno financeiro ao longo do tempo.

```python
try:
    capital = float(input("Capital inicial (R$): "))
    taxa_anual = float(input("Taxa de juros anual (ex: 10 para 10%): ")) / 100
    anos = int(input("Tempo de investimento (em anos): "))
    
    # Fórmula: M = C * (1 + i)**t
    montante = capital * ((1 + taxa_anual) ** anos)
    
    print(f"\nValor investido: R$ {capital:.2f}")
    print(f"Valor Total Final: R$ {montante:.2f}")

except ValueError:
    print("Erro: Digite apenas números válidos usando ponto para decimais.")
```

### Exemplo 2: Criando uma tabela justificada
Podemos usar métodos de string (`ljust`, `center`, `rjust`) para alinhar textos preenchendo o espaço vazio. Ótimo para relatórios no terminal.

`ljust`: ajusta para a esquerda

`center`: ajusta para o centro

`rjust`: ajusta para a direita

```python
print("-" * 50)
print("ATIVO".ljust(20) + "QUANTIDADE".center(15) + "PREÇO (R$)".rjust(15))
print("-" * 50)

# Inserindo os dados formatados
print("ITUB4".ljust(20) + "100".center(15) + f"{34.50:.2f}".rjust(15))
print("VALE3".ljust(20) + "50".center(15) + f"{68.90:.2f}".rjust(15))
print("-" * 50)
```
