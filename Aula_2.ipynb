# 2.1 Listas 

Armazenando vários valores juntos

---

Para guardar vários valores e variáveis juntos, utilizamos as listas! Determinadas por colchetes [], separando cada elemento com vírgulas

````python
a = 1
b = 2
lista = [a, 3, b]
print(lista)
````

Podemos ter elementos de vários tipos na lista!

````python
a = "Palmeiras"
b = 10
c = False

lista = [a, b, c]
print(lista)
````

## 2.1.1 Índices

---

Para representar cada elemento da lista, denominamos cada elemento com um número. O primeiro item da lista recebe 0, o segundo 1, o terceiro 2, etc...

Para conseguir um item específico da lista, colocamos esse índice entre colcheias ao lado do nome da lista

````python
lista = ["Palmeiras", "Vasco", "Cruzeiro"]

print(lista[0])
print(lista[2])
````

**Cortando listas**

---

Podemos conseguir sublistas da lista original cortando ela por seus índices.

lista[i:j] retorna a sublista que inicia em i e termina em j-1 da lista original

````python
lista = [0, 1, 2, 3]

sublista_02 = lista[0:2]

sublista_23 = lista[2:4]

print(sublista_02)
print(sublista_23)
````

Se i ou j forem vazios, eles assumem valor 0 e n, respectivamente, sendo n o tamanho da lista original

````python
from re import S
lista = [0, 1, 2, 3]

sublista1 = lista[:2]
sublista2 = lista[2:]

print(sublista1)
print(sublista2)
````

## 2.1.2 Adicionando itens à lista

---

Para adicionar um item à lista, não podemos apenas utilizar uma soma, temos que utilizar a função ```append()```

````python
lista = [1, 2, 3]

# lista = lista + 4 # Isto dá erro

lista.append(4)
print(lista)
````

Perceba que podemos ter listas dentro de listas

````python
lista1 = [1, 2, 3]
lista2 = [True, False, True]

lista1.append(lista2)
print(lista1)
````

Agora, se temos duas listas e queremos criar uma lista com todos os elementos das duas listas, aí sim podemos utilizar a soma!

````python
lista1 = [1, 2]
lista2 = [3, 4]

lista3 = lista1 + lista2
print(lista3)
````

Também podemos fazer essa soma para adicionar itens à uma lista

````python
lista1 = [1, 2]
lista2 = [2, 3]

lista1 = lista1 + lista2

print(lista1)

lista1 = lista1 + [4]

print(lista1)
````

## 2.1.3 Removendo elementos da lista

---

Para remover um elemento específico da lista, temos que remover utilizando a função ```remove()``` passando o próprio elemento

````python
lista = [1, 2, 3, 4]

lista.remove(2)

print(lista)
````

É importante ressaltar que, se a lista tiver dois ou mais elementos iguais, ele remove apenas a **primeira aparição** do elemento na lista

````python
lista = ["Maça", "Abacaxi", "Abóbora", "Abacaxi"]

lista.remove("Abacaxi")

print(lista)
````

Se você não souber o elemento que quer remover e sabe apenas o índice dele, é possível removê-lo a partir da função ```pop()```. Além de remover, ele retorna o valor do elemento.

Apenas passe o índice do elemento e, se quiser, guarde o valor do elemento removido em uma variável.

````python
lista = ["Maça", "Abóbora", "Abacaxi"]

item_removido = lista.pop(2)

print(lista)
print(item_removido)
````

Para limpar a lista, utilizamos a função ```clear()```

````python
lista = [1, 2, 3, 4]
lista.clear()
print(lista)
````

# 2.2  Booleanos

Introdução aos valores de tipo booleanos e suas manipulações

---

Você já sabe o que são valores do tipo Booleano (só podem assumir 2 valores), mas sabia que dá pra manipular eles que nem manipulamos int's e float's?

````python
# Variáveis booleanas
verdadeiro = True
falso = False

# f antes da string permite inserir valores de variáveis dentro de {}
mensagem = f"Verdadeiro: {verdadeiro}; Falso: {falso}"

print(mensagem)

print("\nExemplo de manipulação:")
print(not verdadeiro)
print("\nPorque 'not verdadeiro' deu falso???")
````

**Atenção!** Em Python, os valores True e False só são aceitos com a primeira letra maiúscula (True/False), se forem escritos de outra forma (true, FALSE, TrUe, ...) irá dar erro!

## 2.2.1 Representações de Booleanos

---

Já sabemos que os booleanos assumem valores

- verdadeiro: ```True```
- falso: ```False```

Porém, podemos escrevê-los em diferentes forma diferentes do padrão Verdadeiro e Falso. Por exemplo, para números:

- ```1``` = ```True```
- ```0``` = ```False```

Em Python, temos algumas regrinhas para essas representações. Normalmente False equivale ao **vazio** e True representa o contrário disso, ou seja, a presença de algum valor.

Em **Números**: O vazio é 0, que representa False, e qualquer outro número diferente de zero equivale à True

Em **String**: Uma string vazia "" é interpretada como False, qualquer outra string com qualquer caractere é equivalente à True

Em **Listas**: Seguindo a lógica, uma lista é False se for vazia [] e False se tiver qualquer valor dentro


Para transformar qualquer valor para o tipo booleano, basta utilizar o método ```bool()```!


````python
print("Números")
print(f" 0: {bool(0)}")
print(f" 1: {bool(1)}")
print(f"15: {bool(15)}")
print(f"-1: {bool(-1)}") # Valores negativos não são vazios!

print()

print("Strings")
print(f"         \"\":  {bool("")}")
print(f"\"Palmeiras\":  {bool("Palmeiras")}")
print(f"        \" \":  {bool(" ")}") # Espaço é um caractere, " " não é vazio!
````

Essa relação do ```False = vazio``` é verdade para qualquer transformação para boolean, porém, de boolean → número, temos uma exceção:

O Python interpreta ```1 = True``` e ```0 = False```

Isso nos permite realizar operações (que já estamos acostumados) com os próprios booleans

````python
v = True
f = False

print(v + f) # 1 + 0 = 1
print(f - v) # 0 - 1 = -1
print(v * f) # 0 x 1 = 0
````

## 2.2.2 Operadores Booleanos

---

Na matemática, temos os operadores ```+```, ```-``` e ```/``` para manipular números. Para os booleanos, temos: ```and```, ```or``` e ```not```

**Operador ```and``` (```&```)**

---

Em português, é o operador ```e```. Ele combina duas condições e retorna True **apenas se as duas condições são verdadeiras ao mesmo tempo**, se qualquer uma das duas for falsa, o resultado será `False`

````python
verdadeiro = True
falso = False

print(verdadeiro and verdadeiro)
print(verdadeiro and falso)
print(falso and falso)
````

Podemos, também, realizar essa exata mesma operação pelo operador `&`

````python
print(verdadeiro & verdadeiro)
print(verdadeiro & falso)
print(falso & falso)
````

**Operador ```or``` (```|```)**

---

Em português, é o operador ```ou```. Ele verifica duas condições e retorna `True` **se alguma das duas condições for verdadeira**, caso contrário, retorna `False`

````python
verdadeiro = True
falso = False

print(verdadeiro or verdadeiro)
print(verdadeiro or falso)
print(falso or falso)
````

Essa exata mesma operação pode ser realizada pelo operador `|`

````python
print(verdadeiro | verdadeiro)
print(verdadeiro | falso)
print(falso | falso)
````

<div align="center" style="font-size: 22px;">

| variáveis \\ operador | and   | or    |
|:---------------------:|:-----:|:-----:|
| True e True           | True  | True  |
| True e False          | False | True  |
| False e False         | False | False |

</div>


**Operador ```not```**

---

Esse é diferente! 
Em português, é o operador ```não```. Ele verifica uma condição apenas e retorna **exatamente o seu contrário**, ou seja, transforma `True` $\rightarrow$ `False` e `False` $\rightarrow$ `True`

Além disso, ele é escrito **antes** do nome da variável

````python
verdadeiro = True
falso = False

print(verdadeiro)
print(not verdadeiro)
print(falso)
print(not falso)
````

# 2.3  Condicionais

Como escrever perguntas em Python?

---

Uma grande parte da lógica de programação é realizar ações apenas em algumas condições específicas: Multiplicar valores apenas se um número for positivo, imprimir algo apenas se uma variável for True, etc...

Passando para o mundo real, teríamos situações parecidas com essas: Passar o aluno apenas se a sua nota for maior que sete, frete grátis apenas se a compra for maior que R$150, etc...

Percebeu que as quatros situações descritas possuem uma palavra em comum?

## 2.3.1 Condicional ```if```

---

Em português, é o condicinal ```se```. Recebe uma condição e executa uma ação (bloco de código) apenas se essa condição for **válida**.

Por enquanto, iremos pensar a condição apenas como um valor boolean, ou seja, a condição só pode assumir formato de uma pergunta que tem sim (True) ou não (False).

Pensando em booleans, ela recebe uma variável boolean e executa um bloco de código apenas se essa variável for **True**.

> Passe de ano apenas **se** o aluno foi aprovado.

Na estrutura do if (condição → ação):

> Aluno foi aprovado? Se sim, passe o aluno de ano.

**Estrutura do ```if```**

```
if condição:
  Bloco de código
```

Perceba que o bloco de código que pertence ao ```if``` sempre possui uma *indentação* (Espaço/Tabulação) a mais que o if antes.


````python
joao_aprovado = True
maria_aprovado = False

if joao_aprovado:
  print("O aluno joão passou de ano!")

if maria_aprovado:
  print("A aluna maria passou de ano!")
````

Lembra que podemos representar booleans de outras maneiras? Aí que vem a sacada, não necessariamente precisamos escrever um boolean como condição. Basta escrever algo que pode ser representado como boolean!

Exemplo:

> Abre a loja apenas **se** tiver algo no estoque

Em estutura condição → ação:

> Tem algo no estoque? Se sim, abre a loja

Pensando mais em Python:

> O estoque não está vazio? Se sim, abre a loja

Lembra que números equivalem a True apenas se fossem diferente de 0 (vazio)? Então podemos utilizar um número como condição!

````python
quantidade_estoque_restaurante = 120
quantidade_estoque_mercado = 0

if quantidade_estoque_restaurante:
  print("Restaurante aberto!")

if quantidade_estoque_mercado:
  print("Mercado aberto!")
````

## 2.3.2 Condicional ```else```

---

E se quisermos realizar uma ação apenas se a condição não valer? Podemos realizar isso com o condicional ```else```.

Em português, é o condicional ```senão```. Ele complementa o ``ìf``` vindo logo após, realiza uma ação apenas se a condição do ```if```***não*** for válida.

> Realize a consulta apenas **se** o médico está disponível **senão** remarque a consulta

Na estrutura do ```if-else```:

> O médico está disponível? Se sim, realize a consulta. Senão, remarque a consulta

**Estutura do ```if-else```***

```
if condição:
  Bloco de código para condição válida
else:
  Bloco de código para condição inválida

```

````python
dentista_disponivel = True
dermatologista_disponivel = False

if dentista_disponivel:
  print("Realize a consulta do dentista")
else:
  print("Remarque a consulta do dentista")

if dermatologista_disponivel:
  print("Realize a consulta do dermatologista")
else:
  print("Remarque a consulta do dermatologista")
````

````python
quantidade_estoque_restaurante = 120
quantidade_estoque_mercado = 0

if quantidade_estoque_restaurante:
  print("Restaurante aberto!")
else:
  print("Estoque do restaurante vazio!")

if quantidade_estoque_mercado:
  print("Mercado aberto!")
else:
  print("Estoque do mercado vazio!")
````

## 2.3.3 Condicional ```elif```

---

Bem parecido com o ```else```. A diferença é que ele recebe outra condição e realiza a ação apenas se essa nova condição é válida

> Está chovendo? Se sim, leve guarda chuva. Senão, está frio? Se sim, leve o casaco, senão leve o guarda sol.

"*Está chovendo?*" é a condição do ```if```. "*Está frio?*" É a condição do ```elif```

Podemos juntar o ```else``` vindo logo após o ```elif``` para caso a condição do elif for falsa

**Estrutura do ```elif```**

```
if condição1:
  Bloco de código para a condição1 válida
elif condição2:
  Bloco de código para a condição2 válida
else:
  Blocod e código para a condição1 e condição2 inválidas
```


````python
chovendo = False
frio = True

if chovendo:
  print("Leve o guarda chuva")
elif frio:
  print("Leve o casaco")
else:
  print("Leve o guarda sol")
````

````python
quantidade_cozinha = 0
quantidade_estoque = 0

if quantidade_cozinha:
  print("Temos comida na cozinha!")
elif quantidade_estoque:
  print("Temos comida no estoque!")
else:
  print("Não temos comida!")
````

---

# 2.4  Operadores

Vamos fazer perguntas mais complexas!

---

Agora que já entendemos os valores booleanos e as condições, podemos aprender em como comparar variáveis e/ou valores para realizar perguntas mais complexas.

Os operadores que iremos ver são diferentes dos que já aprendemos (+, -, ```and```, /,  ```or```, etc...). Esses **sempre** retornam ```True```ou ```False```, independente se estamos trabalhando com números ou booleans.

## 2.4.1 Operadores de Comparação

---

Percebeu que, até agora, sabemos manipular variáveis e valores mas não compará-los?

Nós não sabemos fazer a pergunta *A variável ```dinheiro``` é positiva?* ou *A variável ```dinheiro``` é maior que 15?*

Os operadores de comparação resolvem isso!

É importante ressaltar que os operadores de comparação trabalham com um tipo de variável por vez, não podemos comparar um número com um boolean, apenas um número com um número ou boolean com boolean

**Igualdade (==)**

---

Tenho certeza que você já está acostumado com esse aqui. É a mesma coisa que comparar se dois valores são iguais, independentemente se são números ou booleans.

**Atenção!** Não confunda ```==``` com ```=```! Eu sei que é confuso, mas lembre que o ```=``` já é utilizado para atribuir um valor à uma variável, então o operador de igualdade não pode utilizar esse símbolo e, portanto, ficamos com o ```==```

````python
int_A = 0
int_B = 7
int_C = 0

print(int_A == 0)
print(int_A == int_A)
print(int_A == int_C)
print(int_B == int_C)
print(int_C == 120)
````

````python
bool_t = True
bool_f = False

print(bool_t == bool_t)
print(bool_t == True)
print(bool_t == bool_f)
print(bool_f == True)
````

````python
string_A = "Palmeiras"
string_B = "Vasco"

print(string_A == "Sport")
print(string_A == string_B)
print(string_B == string_B)
print(string_B == "Vasco")
````

**Diferença (!=)**

---

O contrário da igualdade, verifica se duas variáveis/valores são diferentes. Se ```==``` é True, então ```!=```é False. Se tiver dificuldade, lembre que ```!=``` equivale ao ≠

````python
string_A = "Palmeiras"
string_B = "Vasco"

print(string_A == string_B)
print(string_A != string_B)
print(string_A != "Palmeiras")
print()
print(10 != 14)
print(True != False)
print(5!=5)
````

**Maior/Menor que (<. >, <=, >=)**

---

Esse daqui você provavelmente já conhece. Serve especificamente para comparar apenas números:

- num1 > num2: Número 1 é maior que o número 2
- num1 < num 2: Número 2 é maior que o número 1

Lembra que a boca sempre aponta para o maior número

Também temos o maior/menor que **OU** igual que possuem um símbolo um poco diferente dos que já estamos acostumados:

- \>= equivale ao ≥
- \<= equivale ao ≤

- num1 >= num2: Número 1 é maior ou igual que o número 2
- num1 <= num2: Número 2 é menor ou igual que o número 1

````python
num1 = 0
num2 = 5

print(num1 > num2)
print(num1 < num2)

print()

print(num2 >= 5)
print(num2 <= 3)
print(num2 > 5)
````

## 2.4.2 Operadores de Identidade

---


Servem para verificar se duas variáveis apontam para o mesmo objeto. **Não serve para valores**, apenas para variáveis

Muito complicado? Pense que, quando você faz ```v = 7```, na realidade, você está fazendo uma variável (v) apontar para um objeto (7), a → 7

É importante lembrar que, para listas, cria-se um novo objeto para cada variável (mesmo se as listas forem iguais)

**IS**

---

Retornar True se as duas variáveis apontam para o mesmo objeto, Falso caso contrário.

Para números, os objetos são os próprios números com discernimento entre int e float:

- ```a = 1``` e ```b=1```: As variáveis ```a``` e ```b```apontam para o mesmo objeto (1)

- ```c = 1``` e ```d=1.0```: As variáveis ```a``` e ```b``` não apontam para o mesmo objeto pois ```1``` é um objeto int e ```1.0``` é um objeto float

````python
a = 1
b = 5
c = 1
d = 1.0

f = a

print(a is a)
print(a is b)
print(a is c)
print(a is f)

print("\nDiferença entre int (a) e float (d):")
print(a is d)
print(a == d)
````

````python
a = "Palmeiras"
b = "Vasco"
c = "Palmeiras"


print(a is b)
print(a is c)
````

````python
a = ["Palmeiras", "Vasco"]
b = ["Palmeiras", "Vasco"]
c = a
d = "Vasco"

print(a == b)
print(a is b)
print(a is c)
print(a[0] is d)
print(a[1] is d)
````

**IS NOT**

---

O contrário de ```is```, literalmente

````python
a = 1
b = 5

print(a is b)
print(a is not b)
print(a is not a)
````

## 2.4.2 Operadores de Pertencimento

---

Esses operadores verificam se um valor está presente em uma coleção. Interprete coleção como qualquer junção de valores (string com caracteres, listas com valores, etc...)

**Atenção!*** Para strngs (str), ele diferencia maiúsculas de minúsculas. "brasil" e "Brasil" **não** são o mesmo valor!

**IN**

----

O operador in verifica se um determinado valor/variável está presente em uma coleção

````python
time_grande = "Palmeiras"
a = "meiras"


print("Corinthians" in time_grande)
print("Palmeiras" in time_grande)
print(a in time_grande)
print("Meiras" in time_grande)
````

````python
a = "Palmeiras"
times_grandes = [a, "Flamengo", "Cruzeiro"]

print(a in times_grandes)
print("Palmeiras" in times_grandes)
print("Corinthians" in times_grandes)
````

Para fazer o contrário de ```in``` basta inverter o resultado da opera~çao utilizando o operador ```not```

````python
a = "a"
b = "casa"

print(a in b)
print(not (a in b))
print (not (a in "Palmeiras"))
````

# 2.5  Loops

Rodando entre diversos valores

---

Os loops servem para iterar entre valores, ou seja, passar por vários valores executando a mesma ação.

Serve para evitar trabalho manual e reduzir o tamanho do código. Será uma das ferramentas que você mais utilizará no dia a dia.

## 2.5.1 While loop

---

O While Loop serve para realizar uma ação repetidamente enquanto uma condição é verdadeira.

Similar ao ```if```, o while executa o bloco de código se a condição é valida porém, ao chegar no final, volta à condição e a verifica novamente.

````python
soma = 0
while soma < 100:
  soma = soma + 1
print(soma)
````

Normalmente, criamos uma variável apenas para o loop, chamada de iterador (ou ```i```) que é incrementada em 1 cada vez que o bloco de código roda.

Assim, podemos limitar a quantidade de vezes que o bloco de código vai rodar.

````python
i = 0
num = 320

while i < 5:
  num = num / 2
  i += 1 # Mesma coisa que i = i + 1

print(num)
````

Mas não precisamos necessariamente seguir isso, podemos ter um i que nã inicia no 0

## 2.5.2 For Loop

---

Pense o For Loop como um While Loop mais sofisticado, com o iterador já embutido nele próprio.

Isso torna a sua utilização muito mais fácil que o while, não precisamos ficar criando variáveis iteradoras para cada loop que criarmos

**Estrutura do For loop**

for variavel_iterador in coleção_a_iterar:
  bloco de código

Percebeu que sempre temos que não precisamos de uma condição? Ele apenas precisa de uma coleção para iterar e pronto! Passa por cada item da coleção

Mas como podemos limitar a quantidade de vezes que o loop vai rodar? O Python possui uma função própria ```range(a,b)``` que retona o intervalo [a, b)

````python
print(list(range(0,10)))
````

Assim, podemos utilizar isso a nosso favor!

````python
soma = 0
for i in range(0, 11):
  soma += 1
print(i)
````

Muito mais fácil, né? Mas lembre que também podemos passar por qualquer outra coleção!

````python
palavra = "Palmeiras"
for letra in palavra:
  print(letra)
````

````python
palavra = ""

for letra in "Palmeiras":
  palavra += letra
  print(palavra)
````

Podemos aninhar também!

````python
for i in range(0,3):
  palavra = ""
  for letra in "palmeiras":
    palavra += letra
    print(palavra)
````

Para aninhamentos, podemos manipular os índices, também

````python
for i in range(1,4):
  palavra = ""
  for letra in "palmeiras":
    palavra += letra * i
    print(palavra)
````

**Quando utilizar?**

Diferentemente do while loop, utilizamos o for loop quando soubermos exatamente a quantidade de loops que iremos realizar ou para iterar sobre listas.

````python
chamada = ["Luis", "Pedro", "Márcio"]

# Iteramos sobre a lista de chamada
for aluno in chamada:
    print(aluno, "está na sala?")
````

````python
print("Tabuada do 9:")

for i in range(1, 11):
    valor = 9 * i
    print(f"9 x {i} = {valor}")
````

## 2.5.3 Break

---

E se você quiser que o loop pare, por exemplo, em uma situação de emergência? Você pode!

Com o comando ```break``` você para o loop instantaneamente, terminando e seguindo para o código logo após o loop

````python
i = 0
chance_de_explodir = 0

while i < 10:
  chance_de_explodir += 10

  if chance_de_explodir > 50:
    print("URGẼNCIA, PARAR O LOOP")
    break
  else:
    print(f"Chance de explodir: {chance_de_explodir}, iterador = {i}")

  i += 1

print("Acabei o loop")
````

Percebeu que ele a condição sempre foi satisfeita? o iterador nunca nem passou do 4!

O mesmo operador serve para o for loop

````python
palavra = ""
for letra in "Palmeiras":
  palavra += letra
  if letra == "i":
    break
print(palavra)
````

Para loops aninhados, ele sempre para o loop mais dentro do aninhamento (o que primeiro que vem antes do ```break``)

````python
i = 1
while i < 4:
  palavra = ""
  for letra in "Palmeiras":
    palavra += letra * i
    print(palavra)
    if letra == "i":
      break
  i += 1
````

Percebeu que ele rodou a árvore da palavra 3 vezes? Só que ele sempre parou no i. Isso porque o break sempre quebrou o for loop, então a palavra nunca conseguiu terminar inteira.

````python
altura = 7
base = 12

for a in range(1, altura):
  linha = ""
  for b in range(0, base+1):
    # Deve preencher intervalo [base/2 - a, base/2 + a]
    if (b+1 > base/2 - a) and (b+1 < base/2 + a):
      linha += "/"
    else:
      linha += " "
  print(linha)
````
